```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Define variables
n = sp.symbols('n')
approximation_stirling = n * sp.log(n) - n

# Function to calculate the probability at k = n/2 using Stirling's approximation
def median_probability_stirling(n_value):
    # Approximate binomial coefficient at k = n/2 using Stirling's approximation
    half_n = n_value // 2
    ln_n_factorial = approximation_stirling.subs(n, n_value)
    ln_half_n_factorial = approximation_stirling.subs(n, half_n)
    ln_diff_factorial = approximation_stirling.subs(n, n_value - half_n)
    
    # Use the binomial coefficient formula
    ln_binomial_coefficient = ln_n_factorial - (ln_half_n_factorial + ln_diff_factorial)
    
    # Convert back to the probability scale by exponentiating
    probability = sp.exp(ln_binomial_coefficient) * (0.5**n_value)
    
    return float(sp.N(probability))

# Calculate probabilities for a range of n values
n_values = range(10, 100, 10)
median_probabilities = [median_probability_stirling(n_val) for n_val in n_values]

# Plot the result
plt.figure(figsize=(10, 6))
plt.plot(n_values, median_probabilities, marker='o', linestyle='-', color='blue')
plt.xlabel('n (Total Number of Outcomes)')
plt.ylabel('Median Probability at k = n/2')
plt.title('Change in Median Probability as a Function of n Using Stirling Approximation')
plt.grid(True)
plt.show()
```
