### How to represent a node in a singly linked list in Python

```python
class ListNode(object):
    def __init__(self, value):
        self.next = None
        self.value = value

    def print_all_nodes(self, head):
        while head is not None:
            print(head.value)
            head = head.next

    def search_by_index(self, head, index):
        if head is None or index < 0:
            return None
        for move_times in range(index):
            head = head.next
            if not head:
                return None
        return head

    # homework
    def search_by_value(self, head, value):
        if not head:
            return None
        current_node = head
        while current_node.value is not None:
            if current_node.value == value:
                return current_node
            current_node = current_node.next
        return None


    """
    Since ListNode inherits from object class which implements the equality comparison by identity comparison
    (Two objects will be considered the same if they are located in the same memory address <==> id(obj1) == id(obj2)),
    n1 and n2, while contains the same content, will have different memory addresses for them since
    they are two independent objects. The correct way to solve this is to define our own equallity comparison logic by
    adding `__eq__` implementation ot ListNode class:
    """
    def __eq__(self, other):
        return isinstance(other, ListNode) and self.value == other.value


    def add_to_index(self, head, index, value):
        if index == 0:
            new_node = ListNode(value)
            new_node.next = head
            return new_node
        else:
            prev = self.search_by_index(head, index - 1)
            if prev is None:
                return head

            new_node = ListNode(value)
            new_node.next = prev.next
            prev.next = new_node
            return head



    # def add_to_index(self, head, index, value):
    #     fake_head = ListNode(None)
    #     fake_head.next = head
    #     prev = self.search_by_index(fake_head, index)
    #     if prev is None:
    #         return head
    #     new_node = ListNode(value)
    #     new_node.next = prev.next
    #     prev.next = new_node
    #     return fake_head.next



    # homework
    def remove_from_index(self, head, index):
        fake_head = ListNode(None)
        fake_head.next = head
        remove_place = self.search_by_index(fake_head, index)
        if remove_place is None or remove_place.next is None:
            return fake_head.next
        remove_node = remove_place.next
        remove_place.next = remove_node.next
        remove_node.next = None
        return fake_head.next
```
