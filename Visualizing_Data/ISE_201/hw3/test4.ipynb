```python
import numpy as np
import matplotlib.pyplot as plt

def stirling_binomial(n):
    # Using Stirling's approximation to calculate binomial coefficient for k = n/2
    if n % 2 != 0:
        return 0  # We can only calculate this when n is even
    n_half = n // 2
    ln_n_fact = n * np.log(n) - n
    ln_n_half_fact = n_half * np.log(n_half) - n_half
    ln_binomial = ln_n_fact - 2 * ln_n_half_fact
    binomial = np.exp(ln_binomial)
    p_median = binomial * (0.5 ** n)
    return p_median

# Plotting the median probability for a range of n values
n_values = range(2, 101, 2)
probabilities = [stirling_binomial(n) for n in n_values]

plt.figure(figsize=(10, 6))
plt.plot(n_values, probabilities, marker='o')
plt.yscale('log')  # Use logarithmic scale due to rapid decrease
plt.title('Median Probability of Binomial Distribution Using Stirling\'s Approximation')
plt.xlabel('Number of Trials (n)')
plt.ylabel('Probability of n/2 Successes')
plt.grid(True)
plt.show()
```
